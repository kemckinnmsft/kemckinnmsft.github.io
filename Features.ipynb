```python
import pandas as pd

# Load the Excel file
file_path = 'Data Map Feature Evidence(1-25).xlsx'
df = pd.read_excel(file_path)

# Print the column names to verify the correct name
print(df.columns)

# Accessing by index instead of the long name, but adjusting for 0-based indexing
column_name =  'Features'

# Extract entries from J26 (which is likely row index 25 in the Excel file,
# but row index 24 in the DataFrame due to 0-based indexing)
# and create new column headers
new_columns = df.loc[24, column_name].split(';') # Changed 25 to 24

# Create a new dataframe with the new columns
new_df = pd.DataFrame(columns=new_columns)

# Iterate through rows J2-J25 (which are likely row indices 1-24 in the DataFrame)
for i in range(1, 25):
    entries = df.loc[i-1, column_name].split(';') # Changed i to i-1 to adjust for 0-based indexing
    row_data = {col: 1 if col in entries else 0 for col in new_columns}
    # Changed .append to pd.concat for compatibility with future pandas versions
    new_df = pd.concat([new_df, pd.DataFrame([row_data])], ignore_index=True)

# Display the new dataframe
print(new_df)
new_df.to_excel('output.xlsx', index=False)
```
